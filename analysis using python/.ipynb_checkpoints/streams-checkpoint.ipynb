```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
streams = pd.read_csv('../raw_data/Streams.csv')
```

```python
# top 20 streamed songs
top20 = streams.head(20)

# Plot the barplot
plt.figure(figsize=(16,6))
sns.barplot(top20, y='Song', x='Streams (Billions)', color="#4c72b0")
```

```python
# most streamed artists
artists = streams.groupby(by='Artist').agg({'Streams (Billions)': 'sum'}).sort_values(by='Streams (Billions)',ascending=False).reset_index()

# Plot the barplot
plt.figure(figsize=(16,25))
sns.barplot(artists, y='Artist', x='Streams (Billions)', color="#4c72b0")
```

```python
# how many hits for each artist
artist_mode = streams.groupby(by='Artist').size().reset_index(name='Count').sort_values(by='Count', ascending=False)

# Plot the barplot
plt.figure(figsize=(16,25))
sns.barplot(artist_mode, y='Artist', x='Count', color="#4c72b0")
```

```python
# sum of streams & total of songs released in each year

streams['Release Date'] = pd.to_datetime(streams['Release Date'], format='%d-%b-%y')

grouped_data = streams.groupby(by=streams['Release Date'].dt.year).agg({'Song': 'count', 'Streams (Billions)': 'sum'})
grouped_data
```
