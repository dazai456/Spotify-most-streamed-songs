```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../raw_data/Features.csv')
```

```python
# Descriptive statistics
df.describe()
```

```python
# calculating CV
df01 = df.drop(columns=['id', 'name', 'key', 'mode'])
df01.std() / df01.mean()
```

```python
# Plot histograms for selected attributes
attributes = ['duration', 'energy', 'loudness', 'speechiness', 'acousticness', 'liveness', 'valence', 'tempo', 'danceability']

plt.figure(figsize=(15, 10))
for i, attribute in enumerate(attributes, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[attribute], kde=True)
    plt.title(f'Distribution of {attribute}')

plt.tight_layout()
plt.show()
```
